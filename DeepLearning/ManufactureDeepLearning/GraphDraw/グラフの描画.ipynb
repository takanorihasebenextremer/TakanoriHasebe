```python
#以下Sympyを用いる
from sympy import *
#xというSymbolの作成
x = Symbol('x')
```

```python
#関数の作成
f = x ** 3 - x ** 2 - x + 3
```

```python
#関数をxで微分
grad = diff(f, x)
```

```python
#微分した関数を因数分解
grad.factor()
```

```python
#因数分解したもので, 0を取る値を出す
zero = solve(Eq(grad, 0), x)
```

```python
zero
```

```python
#関数fに0を代入
f.subs([(x, zero[0])])
```

```python
#関数fに1を代入
f.subs([(x, zero[1])])
```
