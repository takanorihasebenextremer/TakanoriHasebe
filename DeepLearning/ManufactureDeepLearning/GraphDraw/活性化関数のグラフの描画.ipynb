```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
from numpy import *
%matplotlib inline
```

```python
#Sigmoid関数の描画
#plt.figure(figsize=(10,6))
x = np.linspace(-8, 8, 1000)
y = 1/(1+exp(-x))
plt.plot(x, y)
```

```python
#tanh関数の描画
#plt.figure(figsize=(7,3))
x = np.linspace(-8, 8, 1000)
y = tanh(x)
plt.plot(x, y)
```

```python
#tanh関数の描画
#expを用いている
x = np.linspace(-8, 8, 1000)
y = (exp(x)-exp(-x))/(exp(x)+exp(-x))
plt.plot(x, y)
```

```python
#sin関数の描画
#x軸にlinspaceを使用
x = np.linspace(0, 6, 1000)
y = sin(x)
plt.plot(x, y)
```

```python
#sin関数の描画
#x軸にarangeを使用
x = np.arange(0, 6, 0.1)
y = np.sin(x)
plt.plot(x, y)
```

```python
#sin関数, con関数を両方とも描画する
x = np.arange(0, 6, 0.1)
y1 = sin(x)
y2 = cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle = "--", label="cos") #linestyleで破線を指定
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos") #titleの作成
plt.legend() #右上にグラフの説明を追加
```

```python
#ステップ関数の実装

def step_function(x): #入力をnumpyの配列となっているところに注意
    y = x > 0 #bool型に変換された
    return y.astype(np.int)

x = np.linspace(-6, 6, 1000)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) #y軸の範囲指定
```
